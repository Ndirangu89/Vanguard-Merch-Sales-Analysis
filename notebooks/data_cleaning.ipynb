## Handling missing data via statistical imputation using NumPy

```python
# Filling missing values with statistical logic


# Impute missing Buyer Age
# np.nanmedian() is used to calculate the median while ignoring the missing (NaN) values.
median_age = np.nanmedian(df['Buyer Age'])
print(f"Calculated Median Buyer Age: {median_age}")
df['Buyer Age'] = df['Buyer Age'].fillna(median_age)
```

```python
# Impute missing Shipping Charges
# Missing shipping costs are replaced with the baseline median shipping rate.
median_shipping = np.nanmedian(df['Shipping Charges'])
print(f"Calculated Median Shipping Charge: ${median_shipping}")
df['Shipping Charges'] = df['Shipping Charges'].fillna(median_shipping)
```

```python
#  Impute missing Rating
# Ratings are ordinal (1 to 5). Imputing the median preserves the baseline customer sentiment.
median_rating = np.nanmedian(df['Rating'])
print(f"Calculated Median Rating: {median_rating} out of 5.0")
df['Rating'] = df['Rating'].fillna(median_rating)
```

```python
# Impute missing Review text
# filling blanks with a formal placeholder to maintain dataset structure.
placeholder_text = 'No Review Provided'
df['Review'] = df['Review'].fillna(placeholder_text)
print(f"Filled missing Reviews with: '{placeholder_text}'")
```

```python
# Confirming the dataset is complete
print("\nMISSING VALUES AFTER IMPUTATION")
# Checking the specific columns just worked on to prove they are now at 0
columns_checked = ['Buyer Age', 'Shipping Charges', 'Rating', 'Review']
print(df[columns_checked].isnull().sum())

print("\nImputation Complete. Dataset demographics and metrics are now stabilized.")
```

## Outlier detection and handling using statistical techniques (NumPy)

```python
# Handling Invalid Financial Data

print("FILTERING INVALID FINANCIAL ENTRIES")
# Justification: Standard B2C retail cannot have a negative sales price. 
# Negative values represent system errors or unrecorded refunds, which distort revenue totals.
initial_row_count = len(df)
df = df[df['Sales Price'] > 0]

removed_invalid_prices = initial_row_count - len(df)
print(f"Action: Removed {removed_invalid_prices} rows with invalid/negative 'Sales Price'.")
```

```python
# Isolating B2B Wholesale Orders (Quantity)

print("\n DETECTING QUANTITY OUTLIERS USING IQR")

# Justification: Massive bulk orders skew the Average Order Value for standard B2C analytics.
# the Interquartile Range method is used via NumPy to find the statistical boundary for normal orders.

# Calculate the 25th percentile (Q1) and 75th percentile (Q3) using NumPy
Q1 = np.percentile(df['Quantity'], 25)
Q3 = np.percentile(df['Quantity'], 75)
IQR = Q3 - Q1

# Define the upper boundary for normal consumer behavior
upper_bound = Q3 + (1.5 * IQR)

print(f"Calculated Q1 (25th Percentile): {Q1}")
print(f"Calculated Q3 (75th Percentile): {Q3}")
print(f"Calculated Upper Bound for B2C Orders: {upper_bound} units")

# Isolate the outliers just to count them
b2b_outliers = df[df['Quantity'] > upper_bound]
print(f"Action: Identified and removed {len(b2b_outliers)} B2B wholesale outlier records.")

# Apply the filter to keep only standard B2C consumer orders
df_cleaned = df[df['Quantity'] <= upper_bound]
```

```python
print("\nFINAL CLEANED DATASET STATUS")

print(f"Original Row Count (Before any cleaning): {initial_row_count}")
print(f"Final Cleaned Row Count: {len(df_cleaned)}")
print(f"Total Rows Removed: {initial_row_count - len(df_cleaned)}")
print(f"Final Dataset Shape: {df_cleaned.shape}")

print("\nOutlier handling complete.")
```

## Export the finalized dataset for Analysis

```python

# Define the output filename
output_filename = 'vanguard_cleaned_data.csv'

# Export the cleaned DataFrame to a CSV file without the index column
df_cleaned.to_csv(output_filename, index=False)

print(f"SUCCESS: Preprocessed dataset exported as '{output_filename}'")
print("Data Cleaning Phase Complete. Ready for Analysis Workflow.")
```
